```python
#Import relevant modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import acc_hits and acc_fa files as numpy arrays
acc_hits = np.loadtxt('https://raw.githubusercontent.com/FinLZ/NEUR265_Zakas/main/acc_hits_hw5.csv', delimiter = ',')
acc_fa = np.loadtxt('https://raw.githubusercontent.com/FinLZ/NEUR265_Zakas/main/acc_fa_hw5.csv', delimiter = ',')
```

```python
#Return shape of each array
print(np.shape(acc_hits))
print(np.shape(acc_fa))
```

```python
#Create a time variable
time = np.linspace(0,4,8001)
```

```python
#Determine sampling rate
sr = len(time) / 4
sr
```

```python
#Plot first trial of acc_hits and acc_fa in separate subplots
fig, ax = plt.subplots(1,2, figsize = (15,4))
ax[0].plot(time, acc_hits[0])
ax[0].set(xlabel = 'Time', ylabel = 'Voltage')
ax[1].plot(time, acc_fa[0])
ax[1].set(xlabel = 'Time', ylabel = 'Voltage')
```

The dominant oscillation is difficult to determine in these figures, but it seems that delta oscillations are dominant in both acc_hits and acc_fa. Acc_hits seems to have a slightly higher oscillation (3-4 Hz) than acc_fa (2-3 Hz)

```python
#Calculate ERP for acc_hits, place a vertical line at 2 seconds. Place a horizontal line on the 'zero' position
plt.plot(time, acc_hits.mean(0))
plt.vlines(2, min(acc_hits.mean(0)), max(acc_hits.mean(0)), 'k', lw=1.5, color = 'black')
plt.hlines(0, time[0], time[-1], color = 'black')
plt.xlabel('Time')
plt.ylabel('Voltage')
```

Before screen touch we see very little LFP activity. After screen touch (time = 2), we see a significant jump in power. From this we can see there is a consistent boost in LFP activity following a hit.

```python
#Calculate ERP for acc_hits and acc_fa, place a vertical line at 2 seconds. Place a horizontal line on the 'zero' position
plt.plot(time, acc_hits.mean(0))
plt.plot(time, acc_fa.mean(0), color = 'orange')
plt.vlines(2, min(acc_hits.mean(0)), max(acc_hits.mean(0)), 'k', lw=1.5, color = 'black')
plt.hlines(0, time[0], time[-1], color = 'black')
plt.xlabel('Time')
plt.ylabel('Voltage')
```

There are sigificant differences between ERPs for hits and false alarms. With a false alarm there is no change in activity following the screen touch. With a hit there is a significant increase in activity, with the power radically increasing.
